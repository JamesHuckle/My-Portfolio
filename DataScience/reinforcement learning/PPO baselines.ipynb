```python
import gym
import numpy as np
import pandas as pd
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
import os

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = PPO(policy='MlpPolicy', env='LunarLander-v2', learning_rate=0.0001, verbose=2)
```

```python
model.learn(total_timesteps=500000)
```

```python
highscores = []
timesteps = 0

avg_reward = 0
while avg_reward < 280:
    model.learn(total_timesteps=15000, n_eval_episodes=100)
    avg_reward = np.mean([ep['r'] for ep in list(model.ep_info_buffer)])
    print('train runs', len(list(model.ep_info_buffer)))
    highscores.append(avg_reward)
    timesteps += 2048
    print(f'timesteps: {timesteps} avg_reward: {avg_reward}')
```

```python
model.ep_info_buffer, np.mean([ep['r'] for ep in list(model.ep_info_buffer)])
```

```python
pd.Series(highscores).rolling(min_periods=1, window=1).mean().plot()
```

```python
env = model.get_env()
test_highscores = []

for i in range(3):
    done = False
    score = 0
    observation = env.reset()
    
    while not done:
        env.render()
        action, _states = model.predict(observation, deterministic=True)
        next_observation, reward, done, info = env.step(action)
        observation = next_observation
        score += float(reward)
    
    print(score)    
    test_highscores.append(score)
    
env.close()
print(test_highscores)
np.mean(test_highscores)
```
