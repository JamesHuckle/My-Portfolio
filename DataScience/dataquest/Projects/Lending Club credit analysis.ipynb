```python
home = os.path.expanduser('~')
data = pd.read_csv(f"{home}/Dropbox/My-Portfolio/DataScience/Data/loan.csv")
data.head()
```

```python
high_income = data[data['annual_inc'] > 300000].index
data.loc[high_income,'annual_inc'] = 300000
```

```python
data.sort_values('annual_inc', ascending=False)
high_value = data[data['annual_inc'] > 150000]
best_titles = high_value.groupby('emp_title').agg({'annual_inc':['median','count']})
best_titles = best_titles[best_titles['annual_inc']['count'] > 150]
best_titles.columns = ['median_income','count']
best_titles.sort_values('median_income',ascending=False)
```
